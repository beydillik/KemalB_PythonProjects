```python
#veri analiz kütüphaneleri 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


#Görselleştirme
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Enerji verisini import etme (excel dosyasından)
df = pd.read_excel("EnergyData.xlsx")

# ML algoritmasının daha iyi öğrenmesi için bu aşırı değerleri çıkarabilriz.. 
df = df[df['electricity_consumption'] >= 2000]
```

```python
#verimizin son hali...
df
```

```python
#Tahmin kolonu ile diğerlerini ayıralım, Gün sütununun (Day) sonuca etkisi olmadığı için de dataframe den çıkaralım.
from sklearn.model_selection import train_test_split
X = df.drop(['Day', 'electricity_consumption'], axis=1)
y = df['electricity_consumption']

```

```python
# X son hali:
X.head()
```

```python
# y son hali:
y.head()
```

```python
#Train Test Bölümlemeleri Oluşturalım
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
x_train.head(7)
```

```python
y_train.head(7)
```

```python
# Algoritma- DecisionTree
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(criterion='squared_error',splitter='best')
dt.fit(x_train,y_train)
#Tahmin/test adımı
y_predict_dt = dt.predict(x_test)
#tahminleri görelim
y_predict_dt
```

```python
#Modeli servislemede kullanmak için dışarı aktaracağız
import joblib

joblib.dump(dt, "DecisionTree_model.sav")
```
